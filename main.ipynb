```python
import numpy as np
X = data['DATE']
X = np.array(X).reshape(-1,1)
```

```python
y = data['RAIN']
```

```python
from sklearn.model_selection import *
from sklearn.metrics import *
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
grid = {
    "n_estimators":[10,100,200,400,800,1600,125,250,500,1250,2500],
    "max_depth":[None,5,10,15,20,25],
    "max_features":["auto","sqrt",'log2'],
    "min_samples_split":[2,4,6,8,10],
    "min_samples_leaf":[1,2,4,6,8,10],
    'bootstrap':[False,True],
    'warm_start':[False,True]
}
model_rs = RandomForestClassifier()
rs = RandomizedSearchCV(estimator=model_rs,param_distributions=grid,cv=5,n_iter=25,verbose=5)
rs.fit(X_train,y_train)
print(rs.best_estimator_)
print(rs.score(X_test,y_test))
print(rs.best_score_)
```
